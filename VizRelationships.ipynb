```python
data = pd.read_csv('WorldHappiness.csv').dropna()
```

Are there any features within the dataset are highly correlated with happiness or with each other ?

```python
correlation_matrix = data[data.columns[1:]].corr()

print("Correlation matrix:")
print(correlation_matrix)
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title("Correlation Matrix Heatmap of factors related to happiness")
plt.show()
plt.savefig('CorrMatrixHeatmap.png')
```

 1 indicates a perfect positive correlation.
-1 indicates a perfect negative correlation.
 0 indicates no correlation.

Highly correlated factors include "Log GDP per capita" and "Healthy life expectancy at birth" (0.83), indicating that wealthier countries tend to have longer life expectancies. "Life Ladder" has a strong correlation with both "Social support" (0.72) and "Log GDP per capita" (0.79), indicating that robust social networks and stable economies are major drivers of life satisfaction.

"Freedom to make life choices" and "Positive affect" (0.58) show positive associations as well, indicating that having more happy emotions is correlated with being more autonomous. Conversely, there is a negative correlation between "Perceptions of corruption" and "Life Ladder" (-0.45), "Freedom to make life choices" (-0.47), and "Positive affect" (-0.29). This suggests that a higher perception of corruption is associated with a lower quality of life, less freedom, and fewer positive emotions.

It's interesting to note that "Negative affect" has a negative correlation with both "Life Ladder" (-0.35) and "Positive affect" (-0.33), indicating that feeling bad about oneself is linked to less happiness with life and fewer pleasant feelings. On the other hand, "Negative affect" and "Perceptions of corruption" have an upward trend (0.27), indicating that corruption may be a factor in the development of negative emotions. These connections reveal more light on the ways in which many elements interact to affect people's general levels of pleasure and wellbeing.

Below you can see some of the relationships. During the investigation I can conclude that life ladder and GDP are the most correlated with other factors.

```python
# Relationship between Log GDP per capita and Healthy life expectancy at birth

log_GDP = data['Log GDP per capita']
life_exp = data['Healthy life expectancy at birth']

plt.scatter(log_GDP, life_exp)

plt.xlabel("Log GDP per capita")
plt.ylabel("Healthy life expectancy at birth")
plt.title(
    "Relationship between Log GDP per capita and Healthy life expectancy at birth")

plt.savefig('GDPvsLifeExp.png')

plt.show()

print("Hypothesis: Higher GDP per capita may be associated with better healthcare systems and higher life expectancy.")

# Relationship between Life Ladder and Negative Affect

life_ladder = data['Life Ladder']
neg_eff = data['Negative affect']

plt.scatter(life_ladder, neg_eff)

plt.xlabel("Life Ladder")
plt.ylabel("Negative affect")
plt.title(
    "Relationship between Life Ladder and Negative affect")

plt.show()
plt.savefig('LifeLaddervsNegEff.png')

print("Hypothesis: Negative emotions are likely to be inversely correlated with life satisfaction.")

# Relationship between Life Ladder and Perceptions of corruption

perc_pf_corrupt = data['Perceptions of corruption']

plt.scatter(life_ladder, perc_pf_corrupt)

plt.xlabel("Life Ladder")
plt.ylabel("Perceptions of corruption")
plt.title(
    "Relationship between Life Ladder and Perceptions of corruption")

plt.show()
plt.savefig('LifeLaddervsPercOfCorr.png')

print("Hypothesis: Lower perceptions of corruption may correlate with higher life satisfaction.")
```
