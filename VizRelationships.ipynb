```python
data = pd.read_csv('WorldHappiness.csv')
```

Are there any features within the dataset are highly correlated with happiness or with each other ?

```python
correlation_matrix = data[data.columns[1:]].corr()

print("Correlation matrix:")
print(correlation_matrix)
```

```python
# add readable part
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title("Correlation Matrix Heatmap")
plt.show()
plt.savefig('CorrMatrixHeatmap.png')
```

 1 indicates a perfect positive correlation.
-1 indicates a perfect negative correlation.
 0 indicates no correlation.

Below you can see some of the relationships. During the investigation I can conclude that life ladder and GDP are mostly correlated with other factors.

```python
# Relationship between Log GDP per capita and Healthy life expectancy at birth

log_GDP = data['Log GDP per capita']
life_exp = data['Healthy life expectancy at birth']

plt.scatter(log_GDP, life_exp)

plt.xlabel("Log GDP per capita")
plt.ylabel("Healthy life expectancy at birth")
plt.title(
    "Relationship between Log GDP per capita and Healthy life expectancy at birth")

plt.show()
plt.savefig('GDPvsLifeExp.png')

print("Hypothesis: Higher GDP per capita may be associated with better healthcare systems and higher life expectancy.")

# Relationship between Life Ladder and Negative Affect

log_GDP = data['Life Ladder']
life_exp = data['Negative affect']

plt.scatter(log_GDP, life_exp)

plt.xlabel("Life Ladder")
plt.ylabel("Negative affect")
plt.title(
    "Relationship between Life Ladder and Negative affect")

plt.show()
plt.savefig('LifeLaddervsNegEff.png')

print("Hypothesis: Negative emotions are likely to be inversely correlated with life satisfaction.")

# Relationship between Life Ladder and Perceptions of corruption

log_GDP = data['Life Ladder']
perc_pf_corrupt = data['Perceptions of corruption']

plt.scatter(log_GDP, perc_pf_corrupt)

plt.xlabel("Life Ladder")
plt.ylabel("Perceptions of corruption")
plt.title(
    "Relationship between Life Ladder and Perceptions of corruption")

plt.show()
plt.savefig('LifeLaddervsPercOfCorr.png')

print("Hypothesis: Lower perceptions of corruption may correlate with higher life satisfaction.")
```
