```python
import matplotlib.pyplot as plt
import numpy as np
!pip install scikit-learn
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
%store -r european_data
```

```python
european_data.head()
european_data["Country name"].unique()
european_data_cleaned = european_data.dropna()
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    european_data_cleaned[['Log GDP per capita', 'Social support', 'Freedom to make life choices']
                  ],
    european_data_cleaned['Healthy life expectancy at birth'],
    test_size=0.2,
    train_size=0.8,
    random_state=42
)

model = LinearRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

plt.scatter(y_test, y_pred)
plt.plot(y_pred, y_pred, color='magenta')
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.title('Actual vs. Predicted Healthy Life Expectancy at Birth')
plt.show()
```
