```python
import matplotlib.pyplot as plt
%store -r south_america_data
%store -r north_america_data
%store -r european_data
%store -r asian_data
%store -r african_data
%store -r oceanian_data
```

EUROPEAN OUTLIERS PLOT 

```python
plt.figure(figsize=(20, 16))
plt.scatter(european_data['Log GDP per capita'],
            european_data['Healthy life expectancy at birth'], c='blue', alpha=0.5)
plt.title('Scatter plot of Log GDP per capita vs. Healthy life expectancy at birth')
plt.xlabel('Log GDP per capita')
plt.ylabel('Healthy life expectancy at birth')

# Highlight outliers
outliers = european_data[(european_data['Log GDP per capita'] > european_data['Log GDP per capita'].quantile(0.95)) | (
    european_data['Healthy life expectancy at birth'] > european_data['Healthy life expectancy at birth'].quantile(0.95))]
for index, row in outliers.iterrows():
    plt.text(row['Log GDP per capita'],
             row['Healthy life expectancy at birth'], row['Country name'], fontsize=8)

plt.show()
```
